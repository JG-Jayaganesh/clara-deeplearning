```python
indent =['greetings','yes','help','lease','buy','lottery','finance','options','entertainment','performance','personal','travel',
'weekend','outing','kit','anything',
'thanks','bye','about','wassup','anything','no','reveal',
'hobby','opportunity','hassle','change','frequently']
import pandas as pd
SYN =pd.DataFrame()
A =pd.DataFrame()
for i in range(len(indent)): 
    SYN['indents']=indent[i]
    similar_words=model.most_similar(indent[i],topn=50)
    similar= [str(similar_words[i][0]) for i in range (len(similar_words))]
    SYN['Synonyms']=similar
    A=A.append(SYN)
```
